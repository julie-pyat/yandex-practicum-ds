## Предобработка данных

### Переименование столбцов

Для удобства работы с данными приведем к общему стилю и переименуем название столбца cityCenters_nearest.

```python
# замена названия столбца cityCenters_nearest на city_centers_nearest и вывод перечня названий столбцов на экран
df.rename(columns={'cityCenters_nearest': 'city_centers_nearest'}, inplace=True)
df.columns
```

### Обработка пропусков

Проверим данные на наличие пропусков.

```python
# определение доли пропусков в столбцах
df.isnull().mean()
```

Как было отмечено выше, количество столбцов, имеющих пропуски в значениях - 14. Из них нас интересуют только 8. Это столбцы ceiling_height, floors_total, living_area, kitchen_area, balcony, locality_name, city_centers_nearest и days_exposition. Рассмотрим данные столбцы подробнее.

Данные в столбцах ceiling_height (высота потолков) floors_total (количество этажей в доме), living_area (жилая площадь) и kitchen_area (площадь кухни) имеют количественные значения, на данном этапе мы пока оставим пропуски как есть.

Столбец city_centers_nearest (расстояние до центра города) рассмотрим более подробно: отдельно изучим список населенных пунктов, имеющие пропуски в столбце city_centers_nearest, и населенные пункты, с заполненными значениями в столбце.

```python
df.query('city_centers_nearest.isnull()')['locality_name'].value_counts().head(10)
```

```python
df.query('not city_centers_nearest.isnull()')['locality_name'].value_counts()
```

Из полученных данных можно сделать вывод, что пустые строки в столбце city_centers_nearest имеют квартиры, расположенные в Ленинградской области, а квартиры с заполненными данными в данном столбце находятся в населенных пунктах, которые входят в состав агломерации Санкт-Петербурга, за исключением разве что поселка Мурино и Щеглово.

Создадим столбец, где где будем хранить бинарное значение True если это объект расположен в агломерации Санкт-Петерубруга и False - если в Ленинградской области. Так как по Петербургу количество населенных пунктов меньше и их можно вручную проверить, входит ли данный населенный пункт в состав Санкт-Петербурга, сначала заполним только единицы, а после обработки скрытых дубликатов в столбце locality_name, можно будет данные дополнить.

```python
df['is_city'] = ~df['city_centers_nearest'].isnull()
```

Теперь вручную заменим статус города в поселках Мурино и Щеглово, так как они относятся к Ленинградской области.

```python
df.loc[(df.locality_name == 'поселок Мурино'), 'is_city'] = False
df.loc[(df.locality_name == 'посёлок Щеглово'), 'is_city'] = False
```

```python
# проверка
df.query('is_city == True')['locality_name'].value_counts()
```

Пока пустые значения столбца city_centers_nearest трогать не будем.

Теперь перейдем к столбцу balcony. Предположим, что если данные о балконах не внесены, значит их в квартире нет, так как их наличие при продаже всегда является достоинством, то есть повышает стоимость квартиры, и если они есть, то их обязательно указывают. Заполним пустые значения столбца balcony нулями.

```python
df['balcony'] = df['balcony'].fillna(0)
```

Рассмотрим теперь столбец locality_name. Так как определить однозначно населенный пункт на данном этапе нельзя, то пока заполним пустые значения на 'Unknown'. Данные могли потеряться при сбое записи в файл или некорректного ввода данных.

```python
df['locality_name'] = df['locality_name'].fillna('Unknown')
```

Перейдем к столбу days_exposition, где хранятся данные о том, сколько времени данное объявление было размещено на сайте. В столбце пустые значения скорее всего означают, что на момент выгрузки данных эти объявления были размещены на сайте, или переведены в архив, или были помечены как ненадежные, а может это просто обычный технический сбой.

Вычислим какая последняя дата была в данных при сохранении данных в нашу таблицу. Для этого нам понадобится столбец даты публикации first_day_exposition.

Для начала заменим формат столбца first_day_exposition с текстового на числовой.

```python
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format = '%Y-%m-%dT%H:%M:%S')  
```

Теперь создадим новый столбец, куда сохраним дату снятия с публикации объявления.

```python
df['last_day_exposition'] = df['days_exposition'].fillna(0).astype(int).apply(lambda x: pd.Timedelta(x, unit="d"))
df['last_day_exposition'] = df['last_day_exposition'] + df['first_day_exposition']
```

Удалим данные тех строк столбца последнего дня публикации last_day_exposition где неизвестны данные в столбце days_exposition.

```python
df['last_day_exposition'] = df['last_day_exposition'].where(~df.days_exposition.isnull())
df.query('days_exposition.isnull()').loc[:,['first_day_exposition', 'last_day_exposition', 'days_exposition']].head()
```

Выясним самую свежую дату в наших данных.

```python
df['last_day_exposition'].agg(['max', 'min'])
```

```python
df.query('days_exposition.isnull()')['first_day_exposition'].agg(['max', 'min', 'mean', 'median'])
```

Построим гистограмму столбца даты публикации first_day_exposition для всех квартир, имеющих пустые значения в столбце days_exposition.

```python
import numpy as np
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
df.query('days_exposition.isnull()')['first_day_exposition'].hist(bins=50, figsize = (10,5))
plt.title('first_day_exposition')
plt.ylabel('values')
```

Как видно из графика подавляющее большинство квартир, имеющих пропуски в столбце days_exposition, имеют дату публикации объявления в 2019 году, а мы выяснили, что дата последней завершенной сделки - 2019-05-02. Таким образом, можно сделать вывод, что пропуски в столбце days_exposition означают, что объявление еще не сняли с публикации на момент сохранения данных в таблице.
 Пока оставим столбец без изменений.

Рассмотрим теперь столбцы, которые не учавствуют в наших гипотезах. Начнем со столбца is_apartment. Данный столбце имеется слишком большой процент пропусков - больше 88%, поэтому информация в нем непрезентативна.

Рассмотрим столбец airports_nearest, где указана информация о расстоянии до ближайшего аэропорта.

```python
df.query('airports_nearest.isnull()')['locality_name'].value_counts().head(15)
```

И чтоже мы видим: информация о расстоянии до ближайшего аэропорта не указана для объектов, находящихся в Ленинградской области, как и информация о расстоянии до центра города. Скорее всего и для таких объектов будет отсутствовать информация о количестве парков, водоемов и о растоянии до ближайшего парка и водоема.

Давайте в этом убедимся.

```python
# кикие поселки имеют пустые значения в столбце parks_around3000
df.query('parks_around3000.isnull()')['locality_name'].value_counts().head()
```

```python
# кикие поселки имеют пустые значения в столбце ponds_around3000
df.query('ponds_around3000.isnull()')['locality_name'].value_counts().head()
```

Теперь убедимся, что их количество совпадает.

Напишем функцию, которая бы получает на вход высоту потолка и возвращает:
* исходное значение, если высота потолка в диапазоне от 2.4 до 5 метров,

Согласно карте жилищного фонда город Санкт-Петербург делится на районы застройки, для которых существуют свои характеристики. В центре города дома старого фонда с высокими потолками 3-4 метра, далее идут дома "сталинского ампира" также с высокими потолками более 3 метров, после идут уже дома массовой застройки - "хрущевки" с высотой потолков 2,5 метра, потом появляются другие серии с высотой потолков от 2,48 до 2,64 метров.

Из данных карты мы видим, что исторический центр занимает радиус 3500 метров, затем идут "сталинки" радиусом примерно до 7000 метров, потом идут крупноблочные дома с первыми "хрущевками" до 9000 метров, далее идут серийные дома улучшенной планировки до 10000 м, и далее уже обычная массовая застройка.

Вычислим для каждой зоны для объектов, находящихся в Санкт-Петербурге, свою высоту потолков, а для других объектов вычислим среднее значение для каждого города.

Создадим функцию, которая относит квартиры Санкт-Петербурга к определенной категории в зависимости от того, в какой зоне она находится или указывает город, если квартира расположена в другом населенном пункте.

```python
def housing_stock(element, city):
    if city == 'санкт-петербург':
        if element < 3500:
            return 'центр'
        elif 3500 <= element < 7000:
            return 'сталинка'
        elif 7000 <= element < 9000:
            return 'крупноблок'
        elif 9000 <= element < 10000:
            return 'хрущевка'
        else:
            return 'панелька'
    else:
        return city
```

Создадим новый столбец housing_stock, в котором город Санкт-Петербург поделен на зоны.

```python
df['housing_stock'] = df.apply(lambda x: housing_stock(x['city_centers_nearest'], x['locality_short_name']), axis=1)
```

Заполним пустые значения медианными, вычисленные по каждой зоне. Если в каком-то населенном пункте у нас не получится вычислить медианное значение, то заполним его общим медианным значением.

```python
df['ceiling_height'] = df['ceiling_height'].fillna(df.groupby('housing_stock')['ceiling_height'].transform('median'))
df['ceiling_height'] = df['ceiling_height'].fillna(df['ceiling_height'].median())
```

Теперь разделим этаж квартиры на три категории: первый, последний, другой и не определено для пустых ячеек. Для этого создадим функцию, которая на вход принимает номер этажа и количество этажей в доме, а возвращает категорию, к каторой относится данный объект.

```python
def floor_category(floor, floors_total):
    if floor == floors_total:
        return 'последний'
    elif floor == 1:
        return 'первый'
    elif floors_total == -1:
        return 'не определено'
    else:
        return 'другой'
```

### Изучение срезов данных

 Создадим функцию hist_bloxplot, которая принимала бы название столбца, а возвращала бы гистограмму и диаграмму размаха из данных столбца.

```python
# функция hist_bloxplot принимает название столбца column, а возвращает гистограмму и диаграмму размаха
def hist_bloxplot(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    Q4 = df[column].max()
    IQR = Q3 - Q1
    start = Q1 - 4 * IQR 
    finish = Q3 + 4 * IQR

    plt.subplot(1, 2, 1)
    if IQR < 10: 
        df[column].hist(bins='sturges', figsize =(15,5))
    else:
        df[(df[column] > start) & (df[column] < finish)][column].hist(bins='doane', figsize =(15,5))
        
    
    plt.title(column)
    plt.xlabel('values')
    plt.ylabel('frequency')
    
    plt.subplot(1, 2, 2)
    df.boxplot(column = column)
    plt.title(column)
    plt.ylabel('values')
    plt.show()
```

Создадим функицию, которая принимает название столбца и значение, а  возвращает долю от общего числа значения столбца к значениям больше указанного числа.

```python
def ratio_data(column, value):
    return df[df[column] > value][column].count() / df[column].count()
```

Создадим список столбцов, зависимость которых со стоимостью квартиры мы анализировали.

```python
list_of_column = ['rooms', 'floors_category', 'weekday', 'month', 'year']
```

Создадим функцию, которая создает столбчатаю диаграмму взаимосвязи стоимость квартир с интересующим там параметром. Она высовид диаграмму для данных квартир Санкт-Петерубра и центральной части города.

```python
def compare_center_and_all_data(column):
    median_spb = spb_df.groupby(column)['last_price_thousand'].median()
    caregory = spb_df[column].unique()

    index = np.arange(spb_df[column].nunique())  # the x locations for the groups
    width = 0.35       # the width of the bars

    fig, ax = plt.subplots(figsize=(10, 5))
    rects1 = ax.bar(index, median_spb, width, color='c')

    median_spb_center = spb_df.query('city_centers_nearest_km < 8').groupby(column)['last_price_thousand'].median()
    rects2 = ax.bar(index + width, median_spb_center, width)

    # add some text for labels, title and axes ticks
    ax.set_ylabel('values')
    ax.set_title(column)
    ax.set_xticks(index + width / 2)
    ax.set_xticklabels(median_spb_center.index)

    
    plt.show()
```

Выведим столбчатые диаграммы для сравнения для столбцов rooms, floors_category, weekday, month и year.

```python
for column in list_of_column:
    compare_center_and_all_data(column)
```

Из полученных диаграмм мы можем увидеть, что в столбцах rooms, flor_category, years значения коррелируют, а для столбцов month и year имеют свои особенности.
